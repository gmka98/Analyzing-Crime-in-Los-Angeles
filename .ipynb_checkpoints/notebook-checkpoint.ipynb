# crimes.csv

| Column     | Description              |
|------------|--------------------------|
| `'DR_NO'` | Division of Records Number: Official file number made up of a 2-digit year, area ID, and 5 digits. |
| `'Date Rptd'` | Date reported - MM/DD/YYYY. |
| `'DATE OCC'` | Date of occurrence - MM/DD/YYYY. |
| `'TIME OCC'` | In 24-hour military time. |
| `'AREA NAME'` | The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. For example, the 77th Street Division is located at the intersection of South Broadway and 77th Street, serving neighborhoods in South Los Angeles. |
| `'Crm Cd Desc'` | Indicates the crime committed. |
| `'Vict Age'` | Victim's age in years. |
| `'Vict Sex'` | Victim's sex: `F`: Female, `M`: Male, `X`: Unknown. |
| `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |
| `'Weapon Desc'` | Description of the weapon used (if applicable). |
| `'Status Desc'` | Crime status. |
| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need

crimes["HOUR OCC"] = crimes["TIME OCC"].astype(str).str.zfill(4).str[:2].astype(int)
g = sns.countplot(data=crimes, x="HOUR OCC", color="coral")
peak_crime_hour = crimes["HOUR OCC"].value_counts().idxmax()
# Set axis labels and title properly
g.set_xlabel("Hour of Day")
g.set_ylabel("Number of Crimes")
g.set_title("Crime Frequency by Hour")

print("Peak crime hour:", peak_crime_hour)
```

```python
# Filter for night hours using 24-hour format: 22, 23, 0, 1, 2, 3, 4
night_hours = crimes[crimes["HOUR OCC"].isin([22, 23, 0, 1, 2, 3, 4])]

# Group by area name and count the number of crimes
night_crime_counts = night_hours.groupby("AREA NAME").size()

# Sort the counts in descending order
night_crime_counts_sorted = night_crime_counts.sort_values(ascending=False)

# Get the area with the most night crimes
peak_night_crime_location = night_crime_counts_sorted.index[0]

print("Area with most night crimes:", peak_night_crime_location)
```

```python
# Filter out negative age values (if any)
crimes = crimes[crimes["Vict Age"] >= 0]

# Define the age bins and labels
bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']

# Create a new column with age groupings
crimes["Age Bracket"] = pd.cut(crimes["Vict Age"], bins=bins, labels=labels)

# Count crimes per age bracket
victim_ages = crimes["Age Bracket"].value_counts().sort_index()

# Display the result
print(victim_ages)
```
